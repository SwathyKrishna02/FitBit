 ## k-means clustering 

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
activity_features = ['TotalSteps', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories']
```

```python
sleep_features = ['TotalSleepRecords', 'TotalMinutesAsleep', 'TotalTimeInBed']
```

```python
data_for_clustering = pd.merge(daily_activity[activity_features], sleep_day[sleep_features], left_index=True, right_index=True)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_for_clustering)
```

```python
k = 3
```

```python
kmeans = KMeans(n_clusters=k, random_state=42)
data_for_clustering['cluster'] = kmeans.fit_predict(scaled_data)
```

```python
cluster_analysis = data_for_clustering.groupby('cluster').mean()
```

```python
plt.figure(figsize=(12, 6))
cluster_analysis.T.plot(kind='bar')
plt.title('Cluster Characteristics')
plt.xlabel('Features')
plt.ylabel('Mean Values')
plt.show()
```
